```python
import numpy as np
import pandas as pd
from plotnine import ggplot, aes, geom_line, geom_point, labs
```

```python
#Reading the file
df1=pd.read_csv(r"C:\\Set2\\Oxford Climate Data.csv", low_memory=False)
```

```python
(
    ggplot(df1)
    + aes(x="Year", y="Rain_MM", color="Month")
    + labs(
        x="Year",
        y="Rain in MM",
        color="Month",
        title="Rain for the timeframe in different months",
    )
    + geom_point()
)
```

```python
from plotnine import geom_boxplot
ggplot(df1) + aes(x="Month", y="MaxTemp") + geom_boxplot()
```

```python
ggplot(df1) + aes(x="Month", y="MinTemp") + geom_boxplot()
```
